```python
import pandas as pd
from pathlib import Path
data_dir = Path(r'C:\Users\Firefly\Desktop\proteomics_breast_cancer')

file = pd.ExcelFile(Path(data_dir, 'phosphopeptides.xlsx'), engine='openpyxl')
file.sheet_names
```

```python
phospho = file.parse(sheet_name = 'phosphopeptides')
phospho = phospho.loc[:,~phospho.columns.str.contains('normalized')]
phospho = phospho.loc[:,~phospho.columns.str.contains('setB')]
phospho = phospho.iloc[:,:-32]
print(phospho.columns)
```

```python
phospho.rename(columns = {
    'Abundance: setA: 126':'T47D_R1', 
    'Abundance: setA: 127N':'T47D_R2',
    'Abundance: setA: 127C':'BT474_R1',
    'Abundance: setA: 128N':'BT474_R2',
    'Abundance: setA: 129C':'MDA-MB-468_R1',
    'Abundance: setA: 130N':'MDA-MB-468_R2',
    'Abundance: setA: 130C':'MDA-MB-231_R1',
    'Abundance: setA: 131N':'MDA-MB-231_R2',
    'Abundance: setA: 132C':'SUM159PT_R1',
    'Abundance: setA: 133N':'SUM159PT_R2',
    'Abundance: setA: 133C':'MCF10A_R1',
    'Abundance: setA: 134N':'MCF10A_R2',
}, 
               inplace = True     
                        )
phospho = phospho.loc[:, ~phospho.columns.str.contains('Abundance')]
phospho
```

```python
phospho_data = phospho.iloc[:,-12:]
phospho_data
```

```python
phospho_des = phospho.iloc[:,:-12]
phospho_des
```

```python
phospho_proteomics = phospho_data.T
phospho_proteomics = phospho_proteomics.dropna(axis = 1, how = 'any')
phospho_proteomics
```

```python
phospho_proteomics.index = ['T-47D', 'T-47D','BT-474','BT-474','MDA-MB-468','MDA-MB-468','MDA-MB-231','MDA-MB-231','SUM159','SUM159','MCF-10A','MCF-10A']
phospho_proteomics
```

```python
#Normalize: 1. Normalize each data to the sum of that trial
p = phospho_proteomics.copy()
for i in range(phospho_proteomics.shape[0]):
    #p.index[i] = p.index[i][:-3]
    p.iloc[i,:] = p.iloc[i,:]/p.iloc[i,:].sum()
p
```

```python
#Normalize: 2. Normalize across all cell lines (optional)
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import QuantileTransformer
import scipy.stats as spstats

qt = PowerTransformer(method = 'yeo-johnson')
pb = qt.fit_transform(p)

pb= MinMaxScaler(feature_range=(0,1)).fit_transform(pb)
pb= pd.DataFrame(pb)
pb.columns= p.columns
pb.index = p.index
pb
```

```python
pb_mean = pb.groupby(np.arange(len(pb))//2).mean()
pb_mean.index = pb.index.unique()
#pb_mean.to_csv('proteomics_cell_line.csv')
pb_mean
```

```python
phospho_des = phospho_des.loc[phospho_proteomics.columns, :]
phospho_des
```
