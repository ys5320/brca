```python
pca = PCA(n_components=0.9)
new_pca = pd.DataFrame(pca.fit_transform(pb))
importance = pca.explained_variance_ratio_
print(importance)
c=pca.components_

fig=plt.figure(dpi=300)
ax=fig.add_subplot(111)
#color=['lightcoral','indianred','brown','maroon']

for l in pb.index.unique():
    t = new_pca[pb.index == l]
    ax.scatter(t.iloc[:,0],t.iloc[:,1], label = l, marker = 'o',s = 15,alpha = 0.6)
#ax.set_xlim([-1.5,1.5])
#ax.set_ylim([-0.75,1.25])
kwargs={'linestyle':'-', 'lw':0.1, 'width':0.02} 
#for i in np.arange(c.shape[1]):
    #ax.arrow(0,0,c[0,i]*100,c[1,i]*100,alpha=0.5,ec='black',fc='black', **kwargs)
    #ax.text(c[0,i],c[1,i],f'{i+1}', ha="center", va="center",fontsize=16)
plt.xlabel(r'PC 1',fontsize=16)  
plt.ylabel(r'PC 2',fontsize=16)
plt.legend(fontsize=18,frameon=False,loc=[1,0.1])
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.xticks([])
plt.yticks([])
#plt.savefig('combine tsne space.eps',format='eps',dpi=600)
#plt.savefig('combine tsne space.png',dpi=600)
```

```python
h = sns.clustermap(pb, cmap = 'Blues')
```

```python
v1 = pb.iloc[:2,:]
v2 = pb.iloc[2:,:]
v1 = v1.T
v2 = v2.T
v2
```

```python
v_active_means = v1.mean(axis = 1)
v_quiet_means = v2.mean(axis = 1)
v_quiet_means
```

```python
g2_means = v_quiet_means
g1_means = v_active_means
foldchanges = list(np.log2(np.divide(g1_means, g2_means)))

import scipy.stats as stat
pvals = []
num_values = g1_means.shape[0]
g1_values = v1.to_numpy()
g2_values = v2.to_numpy()

#g1_values = g1_means.to_numpy()
#g2_values = g2_means.to_numpy()

for row in range(0, num_values):
    #print(g1_values[row, :],  g2_values[row, :])
    ttest_result = stat.ttest_ind(g1_values[row, :],  g2_values[row, :])
    #ttest_result = stat.ttest_ind(g1_values[row],  g2_values[row])
    pvalue = ttest_result[1]
    pvals.append(pvalue)

#transformed_pvals = list(-1*np.log10(num_values*np.array(pvals)))
transformed_pvals = list(-1*np.log10(np.array(pvals)))
```

```python
d = {'logFC': foldchanges, 'adj.P.Val': transformed_pvals}
df = pd.DataFrame(data = d)
df = pd.concat([des,df], axis = 1)
df
```

```python
plt.figure(dpi = 300)
threshcolor = 'lightgrey'
downcolor = '#2CB8BB'
upcolor = '#E71F19'
plt.scatter(x = foldchanges,
            y = transformed_pvals, 
            s = 5,
            color = threshcolor
           )

# highlight down- or up- regulated genes
p_thresh = -np.log10(0.05)
log_thresh = 2
down = df[(df['logFC']<=-log_thresh)&(df['adj.P.Val']>= p_thresh)]
up = df[(df['logFC']>= log_thresh)&(df['adj.P.Val']>= p_thresh)]

plt.scatter(x=down['logFC'],y=down['adj.P.Val'],s=3,label="Down-regulated",color= downcolor)
plt.scatter(x=up['logFC'],y=up['adj.P.Val'],s=3,label="Up-regulated",color= upcolor)

ddown = down[(down['logFC'] <= -5)|(down['adj.P.Val'] >= 3)]
uup = up[(up['logFC'] >= 5)|(up['adj.P.Val'] >= 3)]

#ddown = down[(down['adj.P.Val'] >= 9)]
#uup = up[(up['adj.P.Val'] >= 2.2)]

n = np.array(ddown['Gene name'])
z = np.array(ddown.logFC)
y = np.array(ddown['adj.P.Val'])
#print(n)
#for i, txt in enumerate(n):
    #plt.annotate(txt, (z[i], y[i]),fontsize = 5)
    
n = np.array(uup['Gene name'])
z = np.array(uup.logFC)
y = np.array(uup['adj.P.Val'])
#rint(n)
#for i, txt in enumerate(n):
    #plt.annotate(txt, (z[i], y[i]),fontsize = 5)
    
    
plt.xlabel("$Log_{2}$FC", fontsize = 14)
plt.ylabel("-$Log_{10}$ P-value", fontsize = 14)
plt.axvline(-log_thresh,color= threshcolor,linestyle="--")
plt.axvline(log_thresh,color= threshcolor,linestyle="--")
plt.axhline(p_thresh,color= threshcolor,linestyle="--")
plt.legend(frameon = False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.savefig(rf'C:\Users\Firefly\Desktop\breast_cancer_figure\single_gene_volcano_MCF10A+TGFB_Theo.png',format = 'png',bbox_inches='tight',dpi=300, transparent = True)
plt.savefig(rf'C:\Users\Firefly\Desktop\breast_cancer_figure\single_gene_volcano_MCF10A+TGFB_Theo.eps',format = 'eps',bbox_inches='tight',dpi=300, transparent = True)
```
