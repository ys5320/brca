```python
# using PLSR
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cross_decomposition import PLSRegression

def vip(model):
    t = model.x_scores_
    w = model.x_weights_
    q = model.y_loadings_
    p, h = w.shape
    vips = np.zeros((p,))
    s = np.diag(t.T @ t @ q.T @ q).reshape(h, -1)
    total_s = np.sum(s)
    for i in range(p):
        weight = np.array([ (w[i,j] / np.linalg.norm(w[:,j]))**2 for j in range(h) ])
        vips[i] = np.sqrt(p*(s.T @ weight)/total_s)
    return vips

X = np.array(pb)
y = np.array([0]*8 + [1]*4)
plsr = PLSRegression(n_components=2)
clf = plsr.fit(X, y)
```

```python
v1 = pb.iloc[:8,:]
v2 = pb.iloc[8:,:]
v1 = v1.T
v2 = v2.T
v2
```

```python
v_active_means = v1.mean(axis = 1)
v_quiet_means = v2.mean(axis = 1)
```

```python
g2_means = v_quiet_means
g1_means = v_active_means
foldchanges = list(np.log2(np.divide(g1_means, g2_means)))
```

```python
import scipy.stats as stat
pvals = []
num_values = g1_means.shape[0]
g1_values = v1.to_numpy()
g2_values = v2.to_numpy()
for row in range(0, num_values):
    ttest_result = stat.ttest_ind(g1_values[row, :],  g2_values[row, :])
    pvalue = ttest_result[1]
    pvals.append(pvalue)
```

```python
transformed_pvals = list(-1*np.log10(num_values*np.array(pvals)))
```
